```python
import requests
import jieba
from bs4 import BeautifulSoup
import pandas as pd

kv = {'user-agent':'Mozilla/5.0'}
```

```python
def is_chinese(uchar):
    """判断一个unicode是否是汉字"""
    if uchar >= u'\u4e00' and uchar <= u'\u9fa5':
        return True
    else:
        return False
```

```python
url = "https://www.booktxt.net/8_8985/"
t = requests.get(url, timeout = 30, headers = kv)
t.encoding = 'gbk'
soup = BeautifulSoup(t.text, 'html.parser')

prefix = 'https://www.booktxt.net/8_8985/'
urls = []
for sin_soup in soup.select("dd > a[href*=.html]"):
    url_t = prefix + sin_soup.get('href')
    urls.append(url_t)
```

```python
non_names = []

for url in urls:
    t = requests.get(url, timeout = 30, headers = kv)
    t.encoding = 'gbk'
    soup = BeautifulSoup(t.text, 'html.parser')

    for sin_soup in soup.select("[id=content]"):
        content = ''
        for ori_text in sin_soup.get_text():

            for cha in ori_text:
                if(is_chinese(cha)):
                    content = content + cha
        
        content = jieba.cut(content)
        
        for non_name in content:
            length = len(non_name)
            if((length >=2) & (length <= 4) & (non_name not in non_names)):
                non_names.append(non_name)
                print(non_name, length)
```

```python
pd.DataFrame(non_names).to_csv('non_names.csv', header = None, index = False)
```
